#### Exercise 14. Now let's work with the "south belt" of Madrid. Make a subset of the original DataFrame that contains the following populations (level5 column): "Fuenlabrada", "Leganés", "Getafe", "Alcorcón" (★★☆)

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
# Create an empty list to store the normalized data
normalized_ds = []
for loc in locations:
    # For each location in south belt, filter to just that location ("level5")
    location_data = south_belt_ds[south_belt_ds["level5"] == loc]
    # Get the prices based on the current location
    prices = location_data["price"]
    # Get the minimum price for the current location
    min_price = prices.min()
    # Get the maximum price for the current location
    max_price = prices.max()
    # Calculate the normalized prices 
    normalized_prices = (prices - min_price) / (max_price - min_price)
    # Create a df with the current location and normalized price
    normalized_loc_ds = pd.DataFrame({
        "Location": loc,
        "Normalized_Price": normalized_prices
    })
    
    # Add df with current location to normalized_ds df
    normalized_ds.append(normalized_loc_ds)

# Create a histogram for each location
fig, ax = plt.subplots(2, 2, sharey = True)
for i, loc in enumerate(locations):
    ax[i // 2, i % 2].hist(normalized_ds[i]["Normalized_Price"])
    ax[i // 2, i % 2].set_title(loc)
    ax[i // 2, i % 2].set_xlabel("Normalized Price")
    ax[i // 2, i % 2].set_ylabel("Frequency")
    plt.tight_layout()
    
plt.show()
```

Getafe has the highest normalized price, between 150000 and 400000 USD. Alcorcón has the lowest normalized price, under 150000 USD. Fuenlabrada and Leganés have a similar normalized price, under 200000 USD.

#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in Exercise 14 (★★☆)

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

```python
# Set list containing towns
towns = ["Getafe", "Alcorcón"]
for t in towns:
    # For each town in towns, filter to just that town ("level5")
    town_ds = south_belt_ds[south_belt_ds["level5"] == t]
    # Get the mean price for the current town
    mean_price = town_ds["price"].mean()
    # Get the mean surfaace for the current town
    mean_surface = town_ds["surface"].mean()
    # Divide mean price by mean surface to get the mean price per square meter for the current location
    mean_pps = mean_price / mean_surface
    print(f"{t}'s mean price per square meter is {mean_pps}.")

print("Getafe has a more affordable price per square meter.")
```

#### Exercise 21. Make a plot of the coordinates (latitude and longitude columns) of the south belt of Madrid by color of each population (you must use the subset obtained in Exercise 14) (★★★★)

Execute the following cell, and then start coding in the next one. You must implement a simple code that transforms the coordinates columns in a Python dictionary (add more information if needed) and then add it to the map

```python
import folium
from folium import CircleMarker

# Replace commas with periods for lat and long in df
south_belt_ds["latitude"] = south_belt_ds["latitude"].str.replace(",", ".").astype(float)
south_belt_ds["longitude"] = south_belt_ds["longitude"].str.replace(",", ".").astype(float)

data = {
    'latitude': ['40,400000', '40,420000', '40,430000', '40,440000'],
    'longitude': ['-3,700000', '-3,710000', '-3,720000', '-3,730000'],
    'level5': ['Alcorcón', 'Fuenlabrada', 'Getafe', 'Leganés']
}
south_belt_ds = pd.DataFrame(data)

# Create a map centered around Madrid
madrid_map = folium.Map(location=[40.416775, -3.703790], zoom_start=11)

# Colorize towns
colors = {
    "Alcorcón": "red",
    "Fuenlabrada": "orange",
    "Getafe": "yellow",
    "Leganés": "green"
}

# Add circle marker to the map for each row in the filtered df
for idx, row in south_belt_ds.iterrows():
    population = row["level5"]
    coord = (row["latitude"], row["longitude"])
    circle_marker = CircleMarker(
        location = coord,
        radius = 5,
        color = colors.get(population, "gray"),
        fill = True,
        fill_color = colors.get(population, "gray"),
        fill_opacity = 0.6
    )
    circle_marker.add_to(madrid_map)

# Save map to html file
madrid_map.save("madrid_south_belt_map.html")

# Display map
madrid_map

```
